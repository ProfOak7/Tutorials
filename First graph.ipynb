Importing the required module: For this tutorial you are going to import the matplotlib.pyplot package as plt (easier to type). You need to import this package by running the cell below, by either hitting the "run" arrow up above, or hitting: Shift + Enter


```python
import matplotlib.pyplot as plt #Make sure to run this command by hitting run or hitting shift + enter
```

Once the package has been imported, you can successfully make graphs using matplotlib... Let's just make a simple line graph by assigning the following coordinates to x and y:

```python
#Set the x and y values
x = [0,1,2,3,4,5,6,7]  
y = [0,1,2,3,4,5,6,7] 
   
```

Now that we have assigned values to x and y, we can then plot x and y using matplotlib (plt), using the following command:

```python
plt.plot(x, y) # plotting the points 
```

That's great, we made our first graph. But what if you want to add x and y axis? Well, we can do that too using matplotlib:

```python
plt.plot(x, y) # plotting the points 
plt.xlabel('time (hours)') # naming the x axis 
plt.ylabel('time (hours)') # naming the y axis   
```

If we want to add a title, we can do that too:

```python
plt.plot(x, y) # plotting the points 
plt.xlabel('time (hours)') # naming the x axis 
plt.ylabel('time (hours)') # naming the y axis  
plt.title('Time plotted on both x and y axis') # give a title to your graph (note: it automatically puts the title on top) 
plt.show()
```

Maybe you want to make a more interesting graph...

```python
# set the x and y values 
x = [0,8,11,18,24,32,36]  
y = [0.07,3.45,2.96,4.69,0.16,3.73,2.88] 

plt.plot(x, y) # plotting the points 
plt.xlabel('time (hours)') # naming the x axis 
plt.ylabel('Ocean tide (ft)') # naming the y axis  
plt.title('Ocean tide over time') # give a title to your graph (note: it automatically puts the title on top) 
plt.show()
```
