# Machine Learning Types:

# Supervised learning

### Types:

##### Regression: Tries to predict any number out of an infinite possibleof numbers using the shape of a line. 



##### Classification: Can be binary, which tries to predict few outputs based upon the location of the data. For example, trying to determine whether a tumor is benign or malignant... Or whether an image is of a cat or a dog...

## Unsupervised Learning

### Types:

##### Clustering algorithm: A type of grouping that can be performed by the machine on its own. This is commonly used in quantitative biology (hierarchical clustering)

##### Anomaly detection: Find unusual data points

##### Dimensionality reduction: Compress data using fewer numbers
