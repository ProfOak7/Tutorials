```python
# import packages
import statistics 
import pandas as pd
import scipy.stats as stats
from scipy.stats import sem
from statistics import mean, stdev
import matplotlib.pyplot as plt
#include if using a jupyter notebook, remove if using a .py file
%matplotlib inline  
```

```python
#Optional - Set figure resolution to 300dpi
import matplotlib as mpl
%matplotlib inline
plt.rcParams['figure.dpi'] = 300
plt.rcParams['savefig.dpi'] = 300
```

```python
# data
Open = [3, 6 ,4, 6]
Closed = [2, 5, 3, 5]
```

```python
# find the mean using the mean() function from the statistics library
Open_mean = mean(Open)
Closed_mean = mean(Closed)

# find the standard deviation using the stdev() function from the statistics library
Open_stdev = stdev(Open)
Closed_stdev = stdev(Closed)

# find the standard error using SciPY
Open_sem = sem(Open)
Closed_sem = sem(Closed)
```

```python
# build a bar plot
plt.bar(['No Spinach', 'Spinach'],[Open_mean, Closed_mean], yerr=[Open_sem, Closed_sem], capsize=10)
plt.xlabel('')
plt.ylabel('Recovery time (days)')
plt.title('Effect of Spinach Consumption on Recovery Time')
plt.savefig('plot.png')
plt.show()
```
