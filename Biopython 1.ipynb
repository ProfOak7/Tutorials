## Blast

```python
import Bio
```

```python
from Bio.Blast import NCBIWWW #imports blast searcher from web
```

```python
fasta_string = open('./Onedrive/documents/sequences/phix.fa').read() #input yourown fasta file
```

Here is how you can search blast from Jupyter or command line - it may take awhile

```python
result_handle = NCBIWWW.qblast("blastn", "nt", fasta_string) #blastn searches nucleotides, #nt is the nucleotide database to seach
```

```python
from Bio.Blast import NCBIXML #import special XML tool
```

```python
blast_record = NCBIXML.read(result_handle) #special method to read the file into XML
```

```python
len(blast_record.alignments) #show me how long the list of alignments is (by default lists top 50)
```

```python
E_VALUE_THRESH = 0.01 
for alignment in blast_record.alignments:
    for hsp in alignment.hsps:
        if hsp.expect < E_VALUE_THRESH:
            print('***Alignment*** ', alignment.title)
            print('sequence: ', alignment.length)
            print('e value: ', hsp.expect)
            print(hsp.query)
            print(hsp.match)
            print(hsp.sbjct)
```

## Seq

### The Seq object essentially combines a Python string with biological methods. For example:

```python
from Bio.Seq import Seq #Seq is a sequence object that can be imported from Biopython
```

```python
my_seq = Seq('GATTACA')
my_seq
```

### General methods:

The Seq object has a number of methods which act just like those of a Python string, for example the find method:

```python
my_seq.find("TAC")
```

```python
my_seq.find("CAG")
```

There is a count method too:

```python
my_seq.count('A')
```

```python
my_seq.count('AC')
```

However, watch out because just like the Python string’s count, this is a non-overlapping count!

```python
"AAAA".count("AA")
```

In some biological situations, you might prefer an overlapping count using count_overlap which would give three for this example.

### Complement and reverse complement

```python
my_seq.complement()
```

```python
my_seq.reverse_complement()
```

```python
print('Reverse complement is: ', my_seq.reverse_complement())
```

```python
print('Reverse complement is: %s' % my_seq.reverse_complement())
```

### Transcription and back transcription (untranscribe)

```python
my_seq.transcribe()
```

```python
my_rna = my_seq.transcribe()
my_rna
```

```python
my_rna.back_transcribe()
```

If you actually do want the template strand, you’d have to do a reverse complement on top:

```python
my_rna.back_transcribe().reverse_complement()
```
